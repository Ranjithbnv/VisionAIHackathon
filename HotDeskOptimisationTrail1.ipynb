```python
# Timer
import time
import json
from ultralytics import YOLO
import cv2
VIDEO_PATH = "/content/drive/MyDrive/CollabData/demo.mp4"
MODEL_PATH = "/content/drive/MyDrive/CollabData/HotDeskOptimisation/runs/train1/weights/best.pt"   # change if needed



model = YOLO(MODEL_PATH)

# find cabin class id
CLASS_NAMES = model.names
CABIN_ID = None
for k,v in CLASS_NAMES.items():
    if v.lower() == "cabin":
        CABIN_ID = k

cap = cv2.VideoCapture(VIDEO_PATH)
fps = cap.get(cv2.CAP_PROP_FPS)
cap.release()

frames_per_5_sec = int(fps * 5)
frame_count = 0

results = model.track(
    source=VIDEO_PATH,
    tracker="bytetrack.yaml",
    persist=True,
    stream=True
)

for r in results:
    frame_count += 1

    current_cabins = set()

    if r.boxes is not None:
        for box in r.boxes:
            cls = int(box.cls[0])
            tid = int(box.id[0])

            if cls == CABIN_ID:
                current_cabins.add(tid)

    if frame_count % frames_per_5_sec == 0:
        video_time = int(frame_count / fps)

        print(f"[{video_time}s] Cabins visible IDs:", list(current_cabins))
        print(f"[{video_time}s] Cabin count:", len(current_cabins))
        print("-"*50)



```
